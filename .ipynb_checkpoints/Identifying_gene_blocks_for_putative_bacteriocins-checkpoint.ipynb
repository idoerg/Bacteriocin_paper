```python
"""
Making a dictionary of all the results where the key is the genome id
and the value is a tuple with the hit protein id, original protein function, and
original bacterioicin id

blast_result_against_bac_db.xml file has the BLAST result of 1294 collected context
genes against the bacteria refseq database
"""

os.chdir('/home/nafizh/ML_works')
from Bio.Blast import NCBIXML
from collections import defaultdict

result_handle = open("blast_result_against_bac_db.xml", 'r')

blast_records = NCBIXML.parse(result_handle)


bac_genome_protein_dic = defaultdict(list)
for i, blast_record in enumerate(blast_records):
    bacteriocin_id = blast_record.query.split("|")[0]
    original_protein_function = blast_record.query.split("|")[-1]
    print i
    #print bacteriocin_id
    #print original_protein_function
    for j, alignment in enumerate(blast_record.alignments): # alignment = Hit_num in blast result
        #print alignment.title
        #print
        #print alignment.title
        for hsp in alignment.hsps:
            if hsp.expect < 0.000001: # less than 10^-6
                #print blast_record.query
                #print alignment.hit_def
                genome_id = alignment.hit_def.split("|")[0]
                protein_id = alignment.hit_def.split("|")[4]
                #print genome_id
                #print protein_id
                bac_genome_protein_dic[str(genome_id)] += [(str(protein_id), str(original_protein_function), 
                                                            str(bacteriocin_id))]
                #print hsp.identities
                #print hsp.expect
                #print hsp.sbjct
                #print hsp.query
            #break
        #break
    #break

print "Done"
```

```python
# getting bac_genome_coord_dic for all .gbk files
# bac_genome_coord_dic is a dictionary where key is the genbank file id
# and the value is a list of tuples where tuples consist of protein id,
# original protein function - the protein that was blasted against, original bacteriocin id,
# coordinates of the protein, and protein strand
os.chdir('/home/nafizh/bacteria_archaea_genbank_files/bacteria/all_gbk_just_files')




bac_genome_coord_dic = defaultdict(list)
strand_dict = {-1: '-', 1 : '+'}

target_dir = '/home/nafizh/bacteria_archaea_genbank_files/bacteria/all_gbk_just_files'
for root, subFolders, files in os.walk(target_dir):
    for fname in files:
        print fname
        gbk_file_name = fname.split(".")[0]
        in_handle = open(fname, 'r')
        for record in SeqIO.parse(in_handle, 'genbank'):
            for feature in record.features:
                if feature.type == 'CDS':
                    if 'protein_id' in feature.qualifiers:
                        protein_id_list = feature.qualifiers.get('protein_id', ['no protein_id'])
                        for protein_id in protein_id_list:
                            if '.' in protein_id:
                                mystart = feature.location.start.position
                                myend = feature.location.end.position 
                                for key, val in bac_genome_protein_dic.iteritems():
                                    if key == gbk_file_name: # IF the .gbk file exists within all keys
                                        for my_tuples in val:
                                            if my_tuples[0] == protein_id:
                                                bac_genome_coord_dic[gbk_file_name] += [(my_tuples[0], my_tuples[1], my_tuples[2],
                                                                                     str(mystart)+'-'+str(myend), 
                                                                                     strand_dict[feature.strand])]
                                        
                                        

print "Done"
```

```python
"""
Writing down all the individual gene blocks
"""

os.chdir('/home/nafizh/ML_works/all_gene_blocks')
strand_dict = {-1: '-', 1 : '+'}

def write_gene_blocks(fname):
    gbk_file_key = fname.split(".")[0]
    os.chdir('/home/nafizh/bacteria_archaea_genbank_files/bacteria/all_gbk_just_files')
    in_handle = open(fname, 'r')
    
    os.chdir('/home/nafizh/ML_works/all_gene_blocks')
    out_handle = open(gbk_file_key+'_gene_block_prot.fa', 'w')
    
    for record in SeqIO.parse(in_handle, 'genbank'):
        for feature in record.features:
            if feature.type == 'CDS':
                if feature.strand == 1:
                    mystart = feature.location.start.position
                    myend = feature.location.end.position
                    if len(compact_sorted_dict[gbk_file_key]['+']) != 0:
                        for tuple_ in compact_sorted_dict[gbk_file_key]['+']:
                            if mystart >= tuple_[0] and myend <= tuple_[1]:
                                out_handle.write(">protein_id:_%s|product:_%s|note:_%s|%s:%d-%d|%s\n%s\n" % 
                                                         (feature.qualifiers.get('protein_id', ['no protein id'])[0],
                                                         feature.qualifiers.get('product', ['no product'])[0],
                                                         feature.qualifiers.get('note', ['no note'])[0],
                                                         record.name, mystart + 1, myend,
                                                         strand_dict[feature.strand], 
                                                         feature.qualifiers.get('translation', ['no translation'])[0]))
                if feature.strand == -1:
                    mystart = feature.location.start.position
                    myend = feature.location.end.position
                    if len(compact_sorted_dict[gbk_file_key]['-']) != 0:
                        for tuple_ in compact_sorted_dict[gbk_file_key]['-']:
                            if mystart >= tuple_[0] and myend <= tuple_[1]:
                                out_handle.write(">protein_id:_%s|product:_%s|note:_%s|%s:%d-%d|%s\n%s\n" % 
                                                         (feature.qualifiers.get('protein_id', ['no protein id'])[0],
                                                         feature.qualifiers.get('product', ['no product'])[0],
                                                         feature.qualifiers.get('note', ['no note'])[0],
                                                         record.name, mystart + 1, myend,
                                                         strand_dict[feature.strand],
                                                         feature.qualifiers.get('translation', ['no translation'])[0]))
                            
                            
    in_handle.close()
    out_handle.close()
                            
target_dir = '/home/nafizh/bacteria_archaea_genbank_files/bacteria/all_gbk_just_files'
for root, subFolders, files in os.walk(target_dir):
    for fname in files:
        print fname
        write_gene_blocks(fname)
        
print "Done"
```
